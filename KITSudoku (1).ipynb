```python
GRID_WIDTH = 9
GRID_HEIGHT = 9
```

```python
import numpy as np
grid = [[1,0,2,9,0,0,0,4,0], #0
        [6,9,0,0,0,0,0,0,3], #1
        [5,0,0,0,1,8,0,7,9], #2
        [0,2,9,0,8,0,0,0,0], #3
        [4,6,1,0,0,0,7,0,0], #4
        [8,7,5,1,0,0,0,9,0], #5
        [2,5,0,3,0,7,1,0,0], #6
        [0,0,4,0,0,0,0,0,0], #7
        [0,0,0,0,0,1,0,0,0]] #8
#Note to self: show how grid indexing works
```

```python
def is_valid_move(grid,row,col,number):
  for x in range(GRID_WIDTH):
    if grid[row][x] == number:
      return False #If in same row we have the same number, its not a valid move
  for x in range(GRID_HEIGHT):
    if grid[x][col] == number:
      return False #If in same column we have the same number, its not a valid move
  corner_row = row - row % 3 #find corner row
  corner_col = col - col % 3 #find corner column

  for x in range(3):
    for y in range(3):
      if grid[corner_row + x][corner_col + y] == number:
        return False #If there is the same number in the same three by three field, it will not be a valid field
  return True

def solve(grid,row,col):
  if col == 9:
    #array is size 9, so last position (last column value = 8).
    if row == 8:
      #if we have gone over the last column and we are at the last row, we have reached the end and sudoku is solved
      return True
    row += 1
    col = 0
    #if we go past last column, we go to the new row and start with 1st column.
    #when we have gone over the last column and the last row,
    #we know we have found the solution cause false has not been returned yet
  if grid[row][col] > 0: #if the current cell has been filled
    return solve(grid,row,col + 1)
  for num in range(1,10): #1,2,3...9
    if is_valid_move(grid,row,col,num):
      #if it is a valid move
      grid[row][col] = num #for this attempt, assume this is correct solution
      if solve(grid,row,col+1):
        return True #if we are at the end of the gird, we return true as the puzzle is solved.
    grid[row][col] = 0 #if this current number is not a valid move, leave it as 0
  return False #no possible solution for this attempt
```

```python

if solve(grid,0,0):
  for i in range(9):
    for j in range(9):
      print(grid[i][j],end=" ")
    print("\n")
else:
  print("no solution for this sudoku")
```
